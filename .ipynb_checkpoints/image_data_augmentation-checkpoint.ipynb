```python
!pip install tensorflow
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img,img_to_array
import numpy as np
```

```python
#load the image
image_path=r"C:\Users\DELL\Downloads\Animal and Bird Dataset\Dog\14.jpg"
image=load_img(image_path)
image_array=img_to_array(image)
image_array=np.expand_dims(image_array,axis=0)
```

```python
import matplotlib.pyplot as plt
plt.imshow(image)
```

```python
#create ImageDataGenerator with desired augmentations
generator=ImageDataGenerator(rotation_range=270)
```

```python
#apply augmentations using flow
augmented_image_batch=generator.flow(image_array)

#get the augmented image
augmented_image=next(augmented_image_batch)[0].astype('uint8')
```

augmented_image=next(augmented_image_batch):
next() is used to get the next item from the iterator augmented_image_batch. Since the batch size is 1, this will retrieve a single augmented image. At this point, augmented_image would coontain a Numpy array representing the augmented image, but it might not be in the correct data type for display. The [0] is used to access the first element of the array.

.astype('uint8'): this is a crucial step that converts the data type of the augmented image to uint8. Images are typically represented using unsigned 8-bit integers (values from 0 to 255). This data type is suitable for displaying images using libraries like Matplotlib. The augmented image might be in a different data type (ex:float32), so this conversion ensure its in the correct format for display.

```python
plt.imshow(augmented_image)
plt.show()
```

```python
data_gen=ImageDataGenerator(
    width_shift_range=0.2,
    height_shift_range=0.2
) #0.2 means 20% of width, the range of width&height_shift_range is from 0 to 1, negative values allow for left shift
```

```python
generator=data_gen.flow(image_array)
```

```python
augmented_image=next(generator)[0].astype('uint8')
plt.imshow(augmented_image)
plt.show()
```

```python
generator=tf.keras.preprocessing.image.ImageDataGenerator(
    brightness_range=(0.,2.)
)
```

```python
#apply augmentations using flow
augmented_image_batch=generator.flow(image_array, batch_size=1)

#get the augmented image
augmented_image=next(augmented_image_batch)[0].astype('uint8')
```

```python
plt.imshow(augmented_image)
plt.show()
```

```python
for i in range(5):
    augmented_image=next(augmented_image_batch)[0].astype('uint8')
    plt.imshow(augmented_image)
    plt.show()
```
