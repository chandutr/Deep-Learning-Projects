```python
import pandas as pd
df=pd.read_csv(r'customer_churn.csv')
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['customerID'], axis=1,inplace=True)
```

```python
df.dtypes
```

```python
df.nunique()
```

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df['TotalCharges'].isnull().sum()
```

```python
df['TotalCharges'].fillna(df['TotalCharges'].mean(), inplace=True)
```

Split the dataset into input features and target column

```python
x=df.drop('Churn',axis=1) 
y=df['Churn'] 
x.head()
```

encoding

```python
!pip install category_encoders
```

```python
df.head()
```

```python
x.columns
```

```python
cols1=[]
for i in x.columns:
    if x[i].dtype=='object':
        cols1.append(i)
```

```python
y=y.apply(lambda i:1 if i=='Yes' else 0)
```

```python
import category_encoders as ce
encoder=ce.OrdinalEncoder(cols1)
x=encoder.fit_transform(x) 
x.head(2)
```

```python
y.head()
```

Scaling the input feature data

```python
from sklearn.preprocessing import MinMaxScaler
cols_to_scale = ['tenure', 'MonthlyCharges','TotalCharges']
scaler = MinMaxScaler()
x[cols_to_scale] = scaler.fit_transform(x[cols_to_scale])
x
```

Divide into training and testing dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
len(x.columns)
```

Build ANN model

```python
!pip install tensorflow
import tensorflow as tf
from tensorflow import keras
```

```python
x_train.shape
```

```python
x_train
```

```python
model=keras.Sequential([
    keras.layers.Dense(19,input_shape=(19,),activation="relu"),
    keras.layers.Dense(8,activation="relu"),

    keras.layers.Dense(1,activation="sigmoid")
])
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred[:10]
```

```python
y_predictions=[]
for i in y_pred:
    if i>=0.5:
        y_predictions.append(1)
    else:
        y_predictions.append(0)

y_predictions[:10]
        
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test,y_predictions))
```

```python
cm=confusion_matrix(y_test,y_predictions)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Truth")
plt.show()
```
