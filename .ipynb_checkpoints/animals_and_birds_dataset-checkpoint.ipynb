```python
import os
dataset_path=r"C:\Users\DELL\Downloads\Animal and Bird Dataset"
categories=os.listdir(dataset_path)
print(categories)
```

```python
img_path=[]
labels=[]
for category in categories:
    folder_path=os.path.join(dataset_path,category)
    for file in os.listdir(folder_path):
        file_path=os.path.join(folder_path,file)
        img_path.append(file_path)
        labels.append(category)
```

```python
img_path[-5:]
```

```python
labels[-5:]
```

```python
len(labels)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(img_path,labels,test_size=0.2,random_state=42)
```

```python
list(enumerate(categories))
```

```python
label_map={}

for id,category in enumerate(categories):
    label_map.update({category:id})
print(label_map)
```

```python
y_train[:5]
```

```python
y_train1=[]
for i in y_train:
    y_train1.append(label_map[i])
```

```python
y_train1[:5]
```

```python
y_test1=[]
for i in y_test:
    y_test1.append(label_map[i])
```

```python
y_test1[:5]
```

```python
from tensorflow.keras.preprocessing.image import load_img,img_to_array
import numpy as np

def preprocess_image(img_path):
    img=load_img(img_path,target_size=(150,150))
    img_array=img_to_array(img)
    img_array=img_array/255.0
    return img_array
```

```python
y=preprocess_image(img_path[0])
y.shape
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.imshow(y)
```

```python
x_train_new=[]
for i in x_train:
    x_train_new.append(preprocess_image(i))
```

```python
x_train_new=np.array(x_train_new)
x_train_new.shape
```

```python
x_test_new=[]
for i in x_test:
    x_test_new.append(preprocess_image(i))
```

```python
x_test_new=np.array(x_test_new)
x_test_new.shape
```

```python
plt.imshow(x_train_new[0])
```

Build a CNN Model

```python
!pip install tensorflow
import tensorflow as tf
from tensorflow import keras

model=keras.Sequential([
    keras.layers.Conv2D(32,kernel_size=(3,3),activation='relu',input_shape=(150,150,3)), #32,26,26
    keras.layers.MaxPooling2D((2,2)), #32,13,13

    keras.layers.Conv2D(64,kernel_size=(3,3),activation='relu'), #32,64,11,11
    keras.layers.MaxPooling2D((2,2)),

    keras.layers.Flatten(),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dropout(0.4),
    keras.layers.Dense(len(categories),activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train_new,np.array(y_train1),epochs=10)
```

```python
model.evaluate(x_test_new,np.array(y_test1))
```
