```python
df.drop(columns=['date'], axis=1,inplace=True)
df.drop(columns=['id'], axis=1,inplace=True)
df.drop(columns=['yr_renovated'], axis=1,inplace=True)
df.drop(columns=['zipcode'], axis=1,inplace=True)
```

```python
df
```

Split the dataset into input features and target column

```python
x=df.drop('price',axis=1)
y=df['price'] 
```

Scaling the input feature data

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

Divide into training and testing dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=0)
```

```python
len(x.columns)
```

Build ANN Model

```python
!pip install tensorflow
import tensorflow as tf
from tensorflow import keras
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
model=keras.Sequential([
    keras.layers.Dense(100,input_shape=(16,),activation="relu"),
    keras.layers.Dense(100,activation="relu"),

    keras.layers.Dense(1,activation="linear")
])
```

```python
model.compile(optimizer="adam",loss="mse",metrics=["R2Score"])
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
```
